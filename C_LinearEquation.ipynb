```python
# Solve a system of linear equations â€“ Homogenous and Non-Homogenous
def rREF(matrix):
    for i in range(len(matrix)):
        leading(matrix, i)
        pivot(matrix, i)
    return matrix

def leading(matrix, i):
    # divide every cell in that row by the number in the spot going to be the
    # leading one
    if matrix[i][i] == 0:
        return
    matrix[i] = [cell / matrix[i][i] for cell in matrix[i]]

def pivot(matrix, i):
    for r, row in enumerate(matrix):
        if r != i:
            scale = -1 * row[i]
            matrix[r] = [
                curr + scale * other for curr, other in zip(matrix[r], matrix[i])
            ]

def displayM(matrix):
    for row in matrix:
        print("[", end="")
        for cell in row:
            print("{0:>7.2f}".format(cell + 0), end=" ")
        print("]")
    print()

d = int(input("Number of test cases to run: "))
for i in range(d):
    m = []
    rowsOfM = int(input("Number of rows of augmented matrix M: "))
    colsOfM = int(input("Number of cols of augmented matrix M: "))
    for i in range(rowsOfM):
        temp = []
        for j in range(colsOfM):
            temp.append(int(input(f"Enter element M{i+1}{j+1}: ")))
        m.append(temp)

    print("Augmented matrix:")
    displayM(m)
    k = rREF(m)
    print("Solutions:-")
    for i in range(rowsOfM):
        print("X" + str(i + 1) + " = " + str(m[i][colsOfM - 1]))

"""
Number of test cases to run: 3
Number of rows of augmented matrix M: 3
Number of cols of augmented matrix M: 4
Enter element M11: 1
Enter element M12: 1
Enter element M13: -1
Enter element M14: 7
Enter element M21: 1
Enter element M22: -1
Enter element M23: 2
Enter element M24: 3
Enter element M31: 2
Enter element M32: 1
Enter element M33: 1
Enter element M34: 9
Augmented matrix:
[   1.00    1.00   -1.00    7.00 ]
[   1.00   -1.00    2.00    3.00 ]
[   2.00    1.00    1.00    9.00 ]

Solutions:-
X1 = 6.0
X2 = -1.0
X3 = -2.0

Number of rows of augmented matrix M: 3
Number of cols of augmented matrix M: 4
Enter element M11: 1
Enter element M12: -2
Enter element M13: 1
Enter element M14: 0
Enter element M21: 0
Enter element M22: 2
Enter element M23: -8
Enter element M24: 8
Enter element M31: -4
Enter element M32: 5
Enter element M33: 9
Enter element M34: -9
Augmented matrix:
[   1.00   -2.00    1.00    0.00 ]
[   0.00    2.00   -8.00    8.00 ]
[  -4.00    5.00    9.00   -9.00 ]

Solutions:-
X1 = 29.0
X2 = 16.0
X3 = 3.0

Number of rows of augmented matrix M: 3
Number of cols of augmented matrix M: 4
Enter element M11: 2
Enter element M12: 8
Enter element M13: 4
Enter element M14: 2
Enter element M21: 2
Enter element M22: 5
Enter element M23: 1
Enter element M24: 5
Enter element M31: 4
Enter element M32: 10
Enter element M33: -1
Enter element M34: 1
Augmented matrix:
[   2.00    8.00    4.00    2.00 ]
[   2.00    5.00    1.00    5.00 ]
[   4.00   10.00   -1.00    1.00 ]

Solutions:-
X1 = 11.0
X2 = -4.0
X3 = 3.0
"""
```
