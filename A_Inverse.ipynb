```python
#A. Inverse of a Matrix
def transposeM(matrix):
    for i in range(len(matrix)):
        for j in range(i+1, len(matrix)):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
    return matrix

def getMMinor(m,i,j):
    return [row[:j] + row[j+1:] for row in (m[:i]+m[i+1:])]

def getMDeternminant(m):
    #base case for 2x2 matrix
    if len(m) == 2:
        return m[0][0]*m[1][1]-m[0][1]*m[1][0]

    determinant = 0
    for c in range(len(m)):
        determinant += ((-1)**c)*m[0][c]*getMDeternminant(getMMinor(m,0,c))
    return determinant

def getMInverse(m):
    determinant = getMDeternminant(m)
    #special case for 2x2 matrix:
    if len(m) == 2:
        return [[m[1][1]/determinant, -1*m[0][1]/determinant],
                [-1*m[1][0]/determinant, m[0][0]/determinant]]

    #find matrix of cofactors
    cofactors = []
    for r in range(len(m)):
        cofactorRow = []
        for c in range(len(m)):
            minor = getMMinor(m,r,c)
            cofactorRow.append(((-1)**(r+c)) * getMDeternminant(minor))
        cofactors.append(cofactorRow)
    cofactors = transposeM(cofactors)
    for r in range(len(cofactors)):
        for c in range(len(cofactors)):
            cofactors[r][c] = cofactors[r][c]/determinant
    return cofactors

d=int(input("Enter the number of test cases : "))
for d in range(0,d):

    m=[]
    # m=[[4,3,2,2],[0,1,-3,3],[0,-1,3,3],[0,3,1,1]]
    # m=[[5,6,6,8],[2,2,2,8],[6,6,2,8],[2,3,6,7]]
    # m=[[0,6,-2,-1,5],[0,0,0,-9,-7],[0,15,35,0,0],[0,-1,-11,-2,-1],[-2,-2,3,0,-2]]
    rowsOfM = int(input("Number of rows of matrix M: "))
    colsOfM = int(input("Number of cols of matrix M: "))
    for i in range(rowsOfM):
        temp = []
        for j in range(colsOfM):
            temp.append(int(input(f"Enter element M{i+1}{j+1}: ")))
        m.append(temp)
    jk = getMInverse(m)
    def display(jk):
        print("Inverse Matrix : ")
        for r in jk:
            print(r)
    display(jk)

'''
Output:-

Enter number of test cases:3
Number of rows of matrix M: 4
Number of cols of matrix M: 4
Enter element M11: 4
Enter element M12: 3
Enter element M13: 2
Enter element M14: 2
Enter element M21: 0
Enter element M22: 1
Enter element M23: -3
Enter element M24: 3
Enter element M31: 0
Enter element M32: -1
Enter element M33: 3
Enter element M34: 3
Enter element M41: 0
Enter element M42: 3
Enter element M43: 1
Enter element M44: 1
Inverse Matrix:
[0.25, -0.0, -0.075, -0.275]
[-0.0, -0.0, -0.1, 0.3]
[-0.0, -0.16666666666666666, 0.13333333333333333, 0.1]
[-0.0, 0.16666666666666666, 0.16666666666666666, -0.0]

Number of rows of matrix M: 4
Number of cols of matrix M: 4
Enter element M11: 5
Enter element M12: 6
Enter element M13: 6
Enter element M14: 8
Enter element M21: 2
Enter element M22: 2
Enter element M23: 2
Enter element M24: 8
Enter element M31: 6
Enter element M32: 6
Enter element M33: 2
Enter element M34: 8
Enter element M41: 2
Enter element M42: 3
Enter element M43: 6
Enter element M44: 7
Inverse Matrix:
[-17.0, -9.0, 12.0, 16.0]
[17.0, 8.75, -11.75, -16.0]
[-4.0, -2.25, 2.75, 4.0]
[1.0, 0.75, -0.75, -1.0]

Number of rows of matrix M: 5
Number of cols of matrix M: 5
Enter element M11: 0
Enter element M12: 6
Enter element M13: -2
Enter element M14: -1
Enter element M15: 5
Enter element M21: 0
Enter element M22: 0
Enter element M23: 0
Enter element M24: -9
Enter element M25: -7
Enter element M31: 0
Enter element M32: 15
Enter element M33: 35
Enter element M34: 0
Enter element M35: 0
Enter element M41: 0
Enter element M42: -1
Enter element M43: -11
Enter element M44: -2
Enter element M45: -1
Enter element M51: -2
Enter element M52: -2
Enter element M53: 3
Enter element M54: 0
Enter element M55: -2
Inverse Matrix:
[-0.15872302158273383, 0.050809352517985615, -0.013129496402877697, -0.14928057553956833, -0.5]
[-0.03147482014388489, -0.06924460431654676, 0.10107913669064748, 0.3273381294964029, 0.0]
[0.013489208633093525, 0.029676258992805755, -0.014748201438848921, -0.14028776978417265, 0.0]
[-0.16366906474820145, -0.16007194244604317, 0.0856115107913669, 0.302158273381295, 0.0]
[0.210431654676259, 0.06294964028776978, -0.11007194244604317, -0.38848920863309355, 0.0]

'''
```
