```python
# To plot the 2-D and 3-D vector on the cartesian plane
import numpy as np
from matplotlib import pyplot as plt

def Plot2Dvec(vec1, vec2):

    arr = np.array([[0, 0, vec1[0], vec1[1]],
                      [0, 0, vec2[0], vec2[1]]])#to take the numpy array


    A, B, C, D = zip(*arr) #zipping the vectors
    plt.figure()#getting the x-y plane figure (cartesian plane)
    plt.ylabel('Y-axis')
    plt.xlabel('X-axis')
    axis = plt.gca() #to get the access of axes 
    axis.quiver(A, B, C, D, angles='xy', scale_units='xy',
              color=['r', 'b'], scale=1) #to create arrow vector
    axis.set_xlim([0, 10])# to set the limits of x
    axis.set_ylim([0, 10]) # to set the limits of y

    plt.grid()
    plt.draw()
    plt.show()


def Plot3Dvec(vec3, vec4):

    vector3 = np.array(vec3)
    vector4 = np.array(vec4)

    fig = plt.figure()
    axes = plt.axes(projection='3d')
    axes.set_xlim(0, 10)
    axes.set_ylim(0, 10)
    axes.set_zlim(0, 10)
    axes.quiver(0, 0, 0, vector3[0], vector3[1], vector3[2]) 
    axes.quiver(0, 0, 0, vector4[0], vector4[1], vector4[2]) 
    plt.show()


Plot2Dvec([9, 6], [3, 1])
Plot3Dvec([4, 5, 2], [1, 3, 7])
Plot3Dvec([3, 5, 4], [5, 7, 2])
```
